Plot the graph for linear system b and write the point of intersection between three lines.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(5,-5,100)
y1 = (3-2*x)/3
y2 = (16-4*x)/6
y3=(3*x-1)/2

plt.plot(x,y1)
plt.text(-5,2,"2x+3y=3", c="blue", rotation=25)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xticks(range(-10,10))
plt.yticks(range(-10,10))
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(c="black")
plt.axvline(c="black")


plt.grid(ls = ":")
plt.show()
```

prompted for a number, check that the imputed value is a number and then check if the
number is in the matrix. If the number is in the matrix, return the number of times that

```python
a = np.random.randint(1,20, size=(4,4))
print(a)

user_input = input("enter a number you want to search in the matrix")
if user_input.isdigit():
  number = int(user_input)
  if number in a:
    count = np.count_nonzero(number==a)
    print(f"{number} is present {count} times in the generated random matrix")
  else:
    print("entered number not present in matrix")
else:
  print("The number you have entered is not valid number")
```

Task 3:
1. Create a matrix �� of order 5 × 5 with random entries. Calculate the rank of the matrix ��
to check the number of linearly independent columns or rows and to check if the matrix is
invertible. What are the other ways to check if the matrix is invertible?

```python
import numpy as np

# Create a 5x5 matrix with random entries
matrix = np.random.rand(5, 5)
print("Random Matrix:")
print(matrix)
# Calculate the rank of the matrix
rank = np.linalg.matrix_rank(matrix)
print("\nRank of the Matrix:", rank)
# Check determinant
determinant = np.linalg.det(matrix)
if determinant != 0:
    print("\nThe matrix is invertible (determinant is non-zero).")
else:
    print("\nThe matrix is not invertible (determinant is zero).")

# Perform row reduction to check invertibility
reduced_row_echelon_form, _ = np.linalg.qr(matrix)
if np.all(reduced_row_echelon_form == np.eye(5)):
    print("The matrix is invertible (row reduced to identity matrix).")
else:
    print("The matrix is not invertible.")
```

2. Generate a matrix �� of order 5 × 5 with random entries and generate a vector (column) �� of
length 5 with random entries. Calculate Matrix-vector product �� = ����. Now, we have a
linear system, i.e. the matrix �� and a vector ��. Solve this Linear System.

```python
import numpy as np

# Generate a 5x5 matrix with random entries
matrix = np.random.rand(5, 5)
print("Matrix A:")
print(matrix)

# Generate a vector of length 5 with random entries
vector = np.random.rand(5, 1)
print("\nVector b:")
print(vector)

# Calculate the matrix-vector product
result = np.dot(matrix, vector)
print("\nMatrix-Vector Product (A * b):")
print(result)

# Solve the linear system Ax = b
solution = np.linalg.solve(matrix, result)
print("\nSolution x:")
print(solution)
```

3. Solve the following system.
8x+3y−2z=9
−4x+7y+5z==15
3x+4y−12z=35

```python
import numpy as np

# Define the matrix A and vector b
A = np.array([[8, 3, -2],
              [-4, 7, 5],
              [3, 4, -12]])

b = np.array([9, 15, 35])

# Solve the linear system Ax = b
solution = np.linalg.solve(A, b)

# Print the solution
print("Solution (x, y, z):", solution)
```
