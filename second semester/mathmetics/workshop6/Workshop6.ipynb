a. Write a recursive function to find the sum of first n natural numbers.

```python

def sum_of_natural_numbers(n):
    if n <= 1:
        return n
    else:
        return n + sum_of_natural_numbers(n - 1)

n = 5
result = sum_of_natural_numbers(n)
print("Sum of first", n, "natural numbers:", result)
```

Write a non-recursive function, called sumDigits(i), that takes a positive integer and returns
the sum of the digits in the integer. i.e. sumDigitsi(16) = 7, sumDigitsi(111) = 3.
Hint: A while-loop is recommended to process the digits.

```python

def sumDigits(i):
    # Initialize sum to store the sum of digits
    sum_of_digits = 0

    # Iterate through each digit of the number
    while i > 0:
        # Extract the last digit
        digit = i % 10
        # Add the digit to the sum
        sum_of_digits += digit
        # Remove the last digit from the number
        i //= 10

    # Return the sum of digits
    return sum_of_digits

# Example usage
print(sumDigits(16))  # Output: 7
print(sumDigits(111)) # Output: 3
```

sumDigitsr) that returns the sum of the digits in a positive integer. i.e. sumDigitsr(42) =
6, sumDigitsr(1234) = 10.

```python

def sumDigitsr(i):
  if i<10:
    return i
  else:
    return i%10 + sumDigitsr(i // 10)
print(sumDigitsr(42))
print(sumDigitsr(1234))
```

d. The listSearch function (as defined in the preamble) is able to find items on a flat list.
However, neither listSearch or the in operator are able to find items in a tree data
structure such as [1,2,[3,[4],5],6,[7,8]]. So write a new recursive Python function that is
able to find items on a tree, such that:
treeSearch(4, [1,2,[3,[4],5],6,[7,8]]) =&gt; True
treeSearch(9, [1,2,[3,[4],5],6,[7,8]]) =&gt; False
Test your solution to ensure it accurately returns True or False.
Hints: the function is likely to be similar to the listSearch, you may need to use the
function isinstance(&lt; object &gt;, list) and some of the boolean algebra, that you learned last
week, may be useful.

```python

def treeSearch(item, tree):
    if isinstance(tree, list):

        for element in tree:
            if treeSearch(item, element):
                return True
    else:
        # If the current node is not a list (leaf), compare it with the desired item
        if tree == item:
            return True

    # Item not found in the entire tree
    return False

# Example usage:
print(treeSearch(4, [1,2,[3,[4],5],6,[7,8]]))
print(treeSearch(9, [1,2,[3,[4],5],6,[7,8]]))


```

```python

def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n-1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n-1, auxiliary, target, source)
n = 3
tower_of_hanoi(n, 'A', 'C', 'B')
```
