	Create a List containing the first 100 terms of the sequence: 0,4,8,12,16,⋯.
Hint: The given sequence is an AP. In the code below, you can see that either we can create the sequence directly inside the for loop or use the sequence function inside the for loop.


```python
#University_Id =2408059
#Student_name = Dipesh Mahato
# Method 1: Generating sequence directly inside the for loop
sequence = []
for i in range(100):
    term = 4 * i
    sequence.append(term)

print(sequence)

# Method 2: Using the sequence function inside the for loop
sequence = [4 * i for i in range(100)]
print(sequence)
```

	Create a List containing the first 20 terms of the sequence: 2,4,8,16,32,⋯.
Hint: The given sequence is a GP. Write your code similar to the above example containing two methods to create the sequence.


```python

# Method 1: Generating sequence directly inside the for loop
sequence = []
current_term = 2
for _ in range(20):
    sequence.append(current_term)
    current_term *= 2

print(sequence)

# Method 2: Using list comprehension with the formula
sequence = [2 * (2 ** i) for i in range(20)]
print(sequence)
```

3.	Write Python code to find the sum of the sequences created in Problem 1 and Problem 2.
Hint: There are two ways to do so: 1) directly sum all terms in the List using built-in function sum, 2) using the summation formulae for the series.


```python

# Problem 1 sequence: 0, 4, 8, 12, 16, ...
sequence_1 = [4 * i for i in range(100)]

# Problem 2 sequence: 2, 4, 8, 16, 32, ...
sequence_2 = [2 * (2 ** i) for i in range(20)]

# Method 1: Directly summing all terms in the list using built-in function sum
sum_sequence_1_method1 = sum(sequence_1)
sum_sequence_2_method1 = sum(sequence_2)

print("Sum of sequence 1 (Method 1):", sum_sequence_1_method1)
print("Sum of sequence 2 (Method 1):", sum_sequence_2_method1)

# Method 2: Using summation formulae for the series
sum_sequence_1_method2 = 4 * sum(range(100))
sum_sequence_2_method2 = 2 * (2 ** 20 - 1)

print("Sum of sequence 1 (Method 2):", sum_sequence_1_method2)
print("Sum of sequence 2 (Method 2):", sum_sequence_2_method2)
```

4.	Write a Python program to find missing numbers from a list.
Input : [1,2,3,4,6,7,10]
Output : [5, 8, 9]


```python

def find_missing_numbers(lst):
    min_num = min(lst)
    max_num = max(lst)
    missing_numbers = []

    for num in range(min_num, max_num + 1):
        if num not in lst:
            missing_numbers.append(num)

    return missing_numbers

input_list = [1, 2, 3, 4, 6, 7, 10]
missing_numbers = find_missing_numbers(input_list)
print("Input:", input_list)
print("Output:", missing_numbers)

```

6.	Write Python code to generate the first 10 rows of Pascal’s Triangle.

```python

def generate_pascals_triangle(rows):
    triangle = []
    for i in range(rows):
        row = [1]  # First element in each row is always 1
        if i > 0:
            prev_row = triangle[-1]  # Get the previous row
            for j in range(1, i):
                # Calculate the value by adding the two values from the previous row
                row.append(prev_row[j - 1] + prev_row[j])
            row.append(1)  # Last element in each row is always 1
        triangle.append(row)
    return triangle

# Generate the first 10 rows of Pascal's Triangle
pascals_triangle = generate_pascals_triangle(10)

# Print the Pascal's Triangle
for row in pascals_triangle:
    print(row)
```

	Create a Python Program that will create a graph of y = f(x) where;
f(x) = 3x^2   
f(x) = 4x - 3
Your graph will need to plot the values of y for the values of x between -5 and +15 for both lines.
To completely address this question you will need to;
Correctly label the created graph with a title and labelled axes.
Use a function to calculate the values of y for a given x value for both lines.
Use plot markers for the graphs plotted points for both lines.
Include a ‘grid’ on the graph.
Present as evidence your code and the created graph with both lines visible.


```python

import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return 3 * x ** 2

def f2(x):
    return 4 * x - 3

# Generate x values
x_values = np.linspace(-5, 15, 100)

# Calculate y values for each function
y_values_f1 = f1(x_values)
y_values_f2 = f2(x_values)

# Plot the graphs
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_f1, label='f(x) = 3x^2', marker='o')
plt.plot(x_values, y_values_f2, label='f(x) = 4x - 3', marker='o')

# Label the axes and add title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of f(x) = 3x^2 and f(x) = 4x - 3')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
