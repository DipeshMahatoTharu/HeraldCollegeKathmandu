Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show x – Time taken in seconds y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;

And add to your scatter() call the ‘c=colors’ argument like so; plt.scatter(x, y, c=colors) What happens?


```python
import matplotlib.pyplot as plt
import random

x= [random.randint(0,50) for _ in range(10)]
y= [random.randint(0,50) for _ in range(10)]

colors=["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]
plt.scatter(x, y,c=colors)

plt.xlabel('Time taken in seconds')
plt.ylabel('Distance reached')
plt.title('Buzzer Game Results')

plt.show()
```

Following data shows the change in population of different animals from 2017 to 2022.
a.	Create a pie chart for the population of each year
b.	Can we show all the pie chart on the single pane? Write the code.


```python
import matplotlib.pyplot as plt

years = ['2017', '2018', '2019', '2020', '2021', '2022']
bear_population = [8, 54, 93, 116, 137, 184]
dolphins_population = [150, 77, 32, 11, 6, 1]
whales_population = [80, 54, 100, 76, 93, 72]

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs = axs.flatten()

for i, year in enumerate(years):
    total_population = bear_population[i] + dolphins_population[i] + whales_population[i]
    sizes = [bear_population[i], dolphins_population[i], whales_population[i]]
    labels = ['Bear', 'Dolphins', 'Whales']

    axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    axs[i].set_title('Population Distribution for ' + year)

plt.tight_layout()

plt.show()
```

d)	Create a list of colors and using the “ plt.bar(x, y, color = {you list here} ) “set the bar chart colors to be the same as the car colors each bar represents.

```python
import matplotlib.pyplot as plt
car_colors = ['Red', 'Silver', 'Blue', 'White', 'Green']
number_seen = [123, 230, 78, 193, 12]
plt.figure(figsize=(10, 6))
plt.bar(car_colors, number_seen, color=['red', 'silver', 'blue', 'black', 'green'])
plt.xlabel('Car Color')
plt.ylabel('Number Seen')
plt.title('Number of Cars Seen by Color')
plt.show()
```

b)	Implement the code “ plt.bar(x, y, width = 0.3)” to plot the bar chart, what is the effect?

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]  # x-coordinates
y = [10, 20, 15, 25, 30]  # Heights

plt.bar(x, y, width=0.3)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Chart with Width 0.3')

plt.show()
```

e)	Create a pie chart of the car data above and add a title.

```python
import matplotlib.pyplot as plt
labels = ['Toyota', 'Ford', 'Chevrolet', 'Honda']
sizes = [15, 10, 20, 12]
myexplode = [0.3, 0, 0, 0]
plt.pie(sizes, labels=labels, explode=myexplode)
plt.title('Car Sales Distribution')
plt.show()
```
