What are the difference between NumPy array and python list?

The difference between NumPy arrar and python list is:
NumPy array are homegenous This consistency allows for efficient mathematical operations.
Lists can hold different data types (heterogeneous). This flexibility can decrease memory efficiency and slow down numerical computations.

NumPy is a specialized package for scientific computing. To use NumPy arrays, We need to import the numpy module explicitly.
Lists are part of Pythonâ€™s built-in data structures, so no additional module import is required.

NumPy arrays are optimized for numerical computations. They allow element-wise operations (e.g., addition, multiplication) without explicit loops.
Lists do not directly support element-wise operations. You need to use loops or list comprehensions for such computations, which can be less efficient.

 Elements of a NumPy array are stored contiguously in memory, ensuring better memory utilization.
Lists may not store elements in contiguous memory locations, leading to memory fragmentation and potentially higher memory consumption.

NumPy arrays are fast and convenient for mathematical operations. They are widely used in scientific computing, data analysis, and machine learning.
Lists are more general-purpose and versatile but may not perform as well for numerical computations.

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 4 * x**2 + 2

# Generate X values using NumPy
X = np.linspace(-10, 10, 400)  # Generating 400 points from -10 to 10

# Calculate Y values using the function
Y = f(X)

# Plot the graph
plt.plot(X, Y, label='f(x) = 4x^2 + 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph of f(x) = 4x^2 + 2')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors
origin = [0, 0, 0]
ax.quiver(origin, origin, origin, U, V, W, color='r')

# Set plot limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title
ax.set_title('3D Vectors')

# Show plot
plt.show()
```

b. Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors
using Matplotlib.
U = np.array([-0.5, 0.5])
V = np.array([1,1])
The two vectors are added as

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np
import matplotlib.pyplot as plt

# Define the two vectors in 2D space
U = np.array([-0.5, 0.5])
V = np.array([1, 1])

# Add the two vectors to create a third vector
W = U + V

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the vectors
ax.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
ax.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
ax.quiver(0, 0, W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W = U + V')

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Vectors in 2D Space')

# Set the limits of the plot
ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])

# Add legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```

c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of
the vector be U = (1.5, 1.5, 2.0).

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np

def calculate_magnitude(vector):
    """
    Calculate the magnitude of a vector.

    Parameters:
    vector (array_like): The input vector.

    Returns:
    float: The magnitude of the vector.
    """
    return np.linalg.norm(vector)

# Example usage:
U = np.array([1.5, 1.5, 2.0])
magnitude_U = calculate_magnitude(U)
print("Magnitude of vector U:", magnitude_U)
```

d. Write a function to calculate the direction cosines of the vector. For example, let the vector be
O = (1.5, 1.5, 2.0).

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np

def calculate_direction_cosines(vector):
    """
    Calculate the direction cosines of a vector.

    Parameters:
    vector (array_like): The input vector.

    Returns:
    array_like: The direction cosines of the vector.
    """
    magnitude = np.linalg.norm(vector)
    direction_cosines = vector / magnitude
    return direction_cosines

# Example usage:
O = np.array([1.5, 1.5, 2.0])
direction_cosines_O = calculate_direction_cosines(O)
print("Direction cosines of vector O:", direction_cosines_O)
```

e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors
be a= (1.5, 1.5, 2.0) and b = (âˆ’1.5, 0.5, 2.0).

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np

def calculate_scalar_product(vector1, vector2):
    """
    Calculate the scalar (dot) product of two vectors.

    Parameters:
    vector1 (array_like): The first input vector.
    vector2 (array_like): The second input vector.

    Returns:
    float: The scalar product of the two vectors.
    """
    return np.dot(vector1, vector2)

# Example usage:
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
scalar_product_ab = calculate_scalar_product(a, b)
print("Scalar product of vectors a and b:", scalar_product_ab)
```

f. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors
be a = (1.5, 1.5, 2.0) and b= (âˆ’1.5, 0.5, 2.0).

```python
#University_Id = 2408059
#Student_name = Dipesh Mahato
import numpy as np

def calculate_vector_product(vector1, vector2):
    """
    Calculate the vector (cross) product of two vectors.

    Parameters:
    vector1 (array_like): The first input vector.
    vector2 (array_like): The second input vector.

    Returns:
    array_like: The vector product of the two input vectors.
    """
    return np.cross(vector1, vector2)

# Example usage:
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
vector_product_ab = calculate_vector_product(a, b)
print("Vector product of vectors a and b:", vector_product_ab)
```
