```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,4,6,8,9]

plt.plot(x,y,marker="o")
plt.title("Hello")
plt.title("marks")
plt.ylabel("people")
plt.grid(color="green", linewidth=1, linestyle="solid", alpha=0.5)
plt.xlim(1,10)
plt.ylim(2,10)
plt.xticks(rotation=45)
plt.style.use("classic")
print(plt.style.available)
```

Task 2:
The following table shows the different vehicles and the distance covered by them in the span
of 5 days. Display the following table in line plot.
a. Plot “days” in X- axis and “distance covered” in Y- axis.
b. Title of the table “bike details in Plot”

```python
import matplotlib.pyplot as plt
bikes=["ENFIELD","HONDA","YAMAHA","KTM"]
days=[1,2,3,4,5]
distance_covered={
    "ENFIELD":[50,40,70,80,20],
    "HONDA":[80,20,20,50,60],
    "YAMAHA":[70,20,60,40,60],
    "KTM":[40,20,30,70,60]
}
for bike in bikes:
  plt.plot(days, distance_covered[bike],label=bikes)
```
